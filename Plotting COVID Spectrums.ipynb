```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Final Year Projects/Dimensionality Reduction based Spectrum Analysis/COVID Preprocessed Spectral Graph.csv")
```

```python
legend_text=['COVID-19', 'Healthy', 'Suspected']
```

```python
# gca stands for 'get current axis'
#ax = plt.rcParams["font.family"] = "Times New Roman"
#ax = plt.rcParams["font.family"] = "serif"
ax = plt.rcParams["font.size"] = 26
ax = plt.gca()
for i in covid:
    covid_col = i
    df.plot(kind='line',x="V0",y=covid_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#79443B',alpha=1)
for j in healthy:
    healthy_col = j
    df.plot(kind='line',x="V0",y=healthy_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#29AB87',alpha=1)
for l in suspected:
    suspected_col = l
    df.plot(kind='line',x="V0",y=suspected_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#007FFF',alpha=1)

ax = plt.xlabel("Wavenumber ($cm^{-1}$)")
ax = plt.xticks(np.arange(400,2200,step=200),fontsize=24)
ax = plt.ylabel("Raman Intensity (AU)")
ax = plt.yticks(fontsize=24)
ax = plt.legend(legend_text,loc=(0.65,0.715))
ax = plt.grid()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Final Year Projects/Dimensionality Reduction based Spectrum Analysis/Preprocessed COVID Data Spectrum.jpeg',dpi=600, bbox_inches='tight')
ax = plt.show()
```

## **Plotting Preprocessed Average COVID Spectrum**

```python
df2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Final Year Projects/Dimensionality Reduction based Spectrum Analysis/COVID Preprocessed Average Spectral Graph.csv")
columns = list(df2.head(0))
columns
columns.pop(0)
for i in columns:
    y = i
    print(y)
covid = columns[0:1]
healthy = columns[1:2]
suspected = columns[2:3]

# gca stands for 'get current axis'
#ax = plt.rcParams["font.family"] = "Times New Roman"
#ax = plt.rcParams["font.family"] = "Gulliver"
ax = plt.rcParams["font.size"] = 26

ax = plt.gca()
for i in covid:
    covid_col = i
    df.plot(kind='line',x="V0",y=covid_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#79443B',alpha=1)
for j in healthy:
    healthy_col = j
    df.plot(kind='line',x="V0",y=healthy_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#29AB87',alpha=1)
for l in suspected:
    suspected_col = l
    df.plot(kind='line',x="V0",y=suspected_col,ax=ax,legend=False,figsize=(12,8),linewidth=2.5,color='#007FFF',alpha=1)

ax = plt.xlabel("Wavenumber ($cm^{-1}$)")
ax = plt.xticks(np.arange(400,2200,step=200),fontsize=24)
ax = plt.ylabel("Raman Intensity (AU)")
ax = plt.yticks(fontsize=24)
ax = plt.legend(legend_text,loc=(0.65,0.715))
ax = plt.grid()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Final Year Projects/Dimensionality Reduction based Spectrum Analysis/Preprocessed Average COVID Data Spectrum.jpeg',dpi=600, bbox_inches='tight')
ax = plt.show()
```
