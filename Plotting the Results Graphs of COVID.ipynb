### Accuracy Metric Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import seaborn as sns
```

```python
# gca stands for 'get current axis'
#ax = plt.gca()
plt.rcParams["font.family"] = "Times New Roman"
#ax = plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 28

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

# Define data values

kpca = [33, 37, 62, 69, 64, 66]
isomap = [46, 52, 58, 60, 56, 57]
mds = [35, 45, 48, 52, 53, 55]
tsne = [53, 55, 57, 60, 61, 60]
umap = [83, 82, 83, 89, 92, 92]

x = [2, 3, 5, 10, 15, 20]


# Plot a simple line chart

plt.plot(x, kpca,  color ='b', label='KPCA', marker = "o",linewidth=2.5, linestyle = 'dotted')
plt.plot(x, isomap, color ='g', label='ISOMAP', marker = "^",linewidth=2.5, linestyle = 'dashed')
plt.plot(x, mds, color ='m', label='MDS', marker = "s",linewidth=2.5, linestyle = 'dashdot')
plt.plot(x, tsne, color ='c', label='tSNE', marker = "d",linewidth=2.5,linestyle = (0, (3, 1, 1, 1)))
plt.plot(x, umap, color ='r', label='UMAP', marker = "P",linewidth=2.5)

plt.axhline(y = 46, color = 'k', linestyle = (0, (5, 1)),linewidth=2.5, label = "Without DR")
plt.xlabel('Dimensions')
plt.ylabel('Accuracy (%)') 
plt.xticks(x,fontsize=24)
plt.yticks(fontsize=24)
#plt.legend(loc=(0.689,0.59),fontsize=20,framealpha=0.4)
plt.legend(loc=(0.689,0.009),fontsize=20,framealpha=0.5)
plt.gcf().text(0.37, 0.30, '900D', fontsize=24,backgroundcolor='white') 
#plt.savefig('COVID Accuracy Metric.jpeg',dpi=600,bbox_inches='tight')
plt.show()
```

### Matthew's Correlation Coefficient (MCC) Graph 

```python
mcc_df = pd.read_csv('COVID MCC Values.csv')
print(mcc_df)
```

```python
# plot grouped bar chart

plt.rcParams["font.family"] = "Times New Roman"
#plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 28
#ax = plt.gca()

mcc_df.plot(x='Algorithm', kind='bar', figsize=(10,8),stacked=False,width=0.8)
plt.xlabel(" ")
plt.ylabel("Matthew's Correlation Coefficient")


plt.xticks(rotation=0,fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc=(0.801,0.007),fontsize=20)
plt.axhline(y=0.19,color='k',linestyle='dashed',linewidth=2.5)
plt.text(-0.42, 0.175, '900D', fontsize=24,backgroundcolor='w')
plt.grid(axis='y')
plt.savefig('COVID MCC metric.jpeg',dpi=600,bbox_inches='tight')
plt.show()
```
